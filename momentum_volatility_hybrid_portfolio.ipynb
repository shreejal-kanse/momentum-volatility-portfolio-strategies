## I. Executive Summary

**1. What This Project Is About**

This project was designed to test three different investing approaches and find out which one could give the best returns while keeping risk under control. We used real monthly return data from 49 U.S. industries going back several decades. These industries include everything from tech and energy to healthcare and consumer goods, giving us a broad view of the market.

We tested three strategies:

- **Strategy A: Momentum Only** – This one looks back at how each industry performed over the last 12 months. It picks the top 10 and splits the money equally between them. The idea is that industries doing well tend to keep doing well, at least for a little while.

- **Strategy B: Inverse Volatility Only** – This strategy doesn’t look at performance at all. Instead, it checks which industries have had the smallest price swings. The more stable an industry, the more money it gets. It’s a way to try to earn steady returns with lower risk.

- **Strategy C: Momentum + Inverse Volatility** – This combines the first two ideas. First, it picks the top 10 industries based on momentum, then gives more weight to the ones that are less volatile. So, it’s chasing performance, but in a cautious way.

Each portfolio was updated monthly. We tracked how much they earned, how risky they were, and how big their worst drops were. The goal was to see if these simple, rule-based methods could beat the more common investing approaches that often just spread money evenly or follow an index.

In the end, we wanted to learn: Can a basic, thoughtful strategy help us grow money while managing the bumps along the way? The answer turned out to be yes, and Strategy C gave us a clear example of how blending momentum with smart risk control can lead to powerful results.

**2. How We Did It**

### The Data

### How Each Strategy Worked

**Strategy A (Momentum)**

- We picked the top 10 and split our money evenly across them.

**Strategy B (Inverse Volatility)**

- We looked at how much each industry’s returns moved over the past year.
- The more stable ones got more of our money; the riskier ones got less.

**Strategy C (Combined)**

- First, pick the 10 industries that performed best in the past year.

- **Sharpe Ratio**: A score that shows how much return we got per unit of risk.
- **Maximum Drawdown**: The biggest drop the strategy had from its highest point.

## III. Results Discussion

**Strategy A: Momentum Only**

By choosing the top-performing sectors according on their historical returns, Strategy A, called the Momentum Only strategy with equal weights, is meant to exploit the durability of industry performance patterns. The implementation starts with a 12-month lookback period determined % change in returns for every industry in the ind49 dataset, hence generating momentum scores. These scores are arranged in decreasing order, and the portfolio includes the top 20 industries. Every chosen sector gets a fair weight of 1/20, so guaranteeing a balanced distribution without privileging any one sector depending on the size of its momentum. By multiplying these weights by the monthly industry returns and aggregating the results, monthly portfolio returns indicate this momentum-driven strategy's success. The results of the backtest show a standard deviation of 5.8168%, an average monthly return of 1.0078%, and a Sharpe ratio of 0.4296. Between 1969 and 2019, a $1 investment grows to $756.45. Though the comparatively large standard deviation implies that the strategy has considerable volatility, which affects its risk-adjusted returns as seen by the small Sharpe ratio, this performance shows that the approach produces strong long-term growth.

The cumulative return graph for Strategy A shows a mostly upward trend, with the value of a $1 investment rising consistently throughout the decades, especially fast after the 1980s. In contrast, the graph shows instances of severe drawdowns, when the portfolio falls precipitously before regaining its losses, the most prominent of which occurred during the financial crisis of 2008. When market circumstances change unexpectedly, momentum techniques might reverse, but they flourish in changing markets. In behavioral finance, the momentum effect argues that sectors with strong recent performance acquire due to investor herding, delayed news responses, or overconfidence in trends. However, when people's attitudes alter, the same bias might cause them to overreact, which is what happens during market downturns when momentum suddenly reverses. While clear and democratic, equal weighting may dilute the strategy's efficacy by evaluating all top 20 industries equally, even if some have far higher momentum. The portfolio may be unable to fully capitalize on positive trends, limiting upside during strong bull markets.

In terms of risk, Strategy A has a standard deviation of 5.8168%, which indicates a high degree of variability in monthly returns. This is one of the factors that contributes to its lower Sharpe ratio in comparison to other strategies. Due to its dependence on a 12-month lookback time for momentum computation, the strategy may not be able to respond fast enough to shifting market dynamics. As a result, the portfolio may end up holding onto sectors that have already reached their peak momentum. The weighting system does not adjust risk, consequently high-momentum sectors with high volatility are assigned the same weight as more stable ones, increasing portfolio risk. Despite these shortcomings, the strategy's final growth of $756.45 show the ability of capturing large upside over the long term, especially during times of prolonged market trends. Strategy A presents a tempting option for growth-oriented investors who are ready to ride out greater volatility. However, risk-based weighting or a shorter, more flexible lookback time might improve its performance by better capturing evolving patterns.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# === Load Data ===
ind49 = pd.read_csv('data_ind49.csv', parse_dates=['Date']).set_index('Date')
ff5 = pd.read_csv('data_ff5.csv', parse_dates=['Date']).set_index('Date')

# Clean and prep
ff5.columns = ff5.columns.str.strip()
ind49.index = ind49.index.to_period('M')
ff5.index = ff5.index.to_period('M')
ind49 = ind49 / 100
ff5 = ff5 / 100
```

```python
# === Parameters ===
lookback_months = 12
top_n = 20

# === Create volatility ===
momentum_scores = ind49.pct_change(periods=lookback_months)
volatility = ind49.rolling(window=lookback_months).std()

# === Create common index ===
common_index = ind49.index.intersection(momentum_scores.index).intersection(volatility.index).intersection(ff5.index)
ind49 = ind49.loc[common_index]
momentum_scores = momentum_scores.loc[common_index]
volatility = volatility.loc[common_index]
ff5 = ff5.loc[common_index]
```

## Strategy A: Momentum Only (Equal Weight)

```python
momentum_returns = []

for date in common_index[lookback_months:]:
    ranked = momentum_scores.loc[date].sort_values(ascending=False)
    selected = ranked.head(top_n).index

    weights = pd.Series(1 / len(selected), index=selected)  # Equal weight
    returns = ind49.loc[date, weights.index]
    ret = (weights * returns).sum()
    momentum_returns.append((date, ret))

# === Convert to DataFrame ===
df_momentum = pd.DataFrame(momentum_returns, columns=['Date', 'MomentumReturn']).set_index('Date')

# === Add RF and Compute Sharpe ===
df_momentum['RF'] = ff5.loc[df_momentum.index, 'RF']
df_momentum['ExcessReturn'] = df_momentum['MomentumReturn'] - df_momentum['RF']

sharpe_ratio = (
    df_momentum['ExcessReturn'].mean() /
    df_momentum['ExcessReturn'].std()
) * np.sqrt(12)
```

```python
# === Print Results ===
print(f"\n Strategy A: Momentum Only")
print(f"Sharpe Ratio: {sharpe_ratio:.4f}")
print(f"Avg Monthly Return: {df_momentum['MomentumReturn'].mean():.4%}")
print(f"Std Dev: {df_momentum['MomentumReturn'].std():.4%}")
```

```python
# === Plot ===
(1 + df_momentum['MomentumReturn']).cumprod().plot(title='Strategy A: Momentum Only (Equal Weight)')
plt.ylabel("Growth of $1")
plt.grid(True)
plt.show()
```

## Strategy B: Inverse Volatility Only

```python
inv_vol_returns = []

for date in common_index[lookback_months:]:
    vol = volatility.loc[date].replace(0, np.nan).dropna()

    # Inverse volatility weights
    inv_vol = 1 / vol
    weights = inv_vol / inv_vol.sum()

    # Calculate portfolio return
    ret = (weights * ind49.loc[date, weights.index]).sum()
    inv_vol_returns.append((date, ret))

# === Convert to DataFrame ===
df_invvol = pd.DataFrame(inv_vol_returns, columns=['Date', 'InvVolReturn']).set_index('Date')

# === Add RF and Compute Sharpe ===
df_invvol['RF'] = ff5.loc[df_invvol.index, 'RF']
df_invvol['ExcessReturn'] = df_invvol['InvVolReturn'] - df_invvol['RF']

sharpe_ratio = (
    df_invvol['ExcessReturn'].mean() /
    df_invvol['ExcessReturn'].std()
) * np.sqrt(12)
```

```python
# === Print Results ===
print(f"\n Strategy B: Inverse Volatility Only")
print(f"Sharpe Ratio: {sharpe_ratio:.4f}")
print(f"Avg Monthly Return: {df_invvol['InvVolReturn'].mean():.4%}")
print(f"Std Dev: {df_invvol['InvVolReturn'].std():.4%}")
```

```python
# === Plot Cumulative Return ===
(1 + df_invvol['InvVolReturn']).cumprod().plot(title='Strategy B: Inverse Volatility Only')
plt.ylabel("Growth of $1")
plt.grid(True)
plt.show()
```

```python
# === Rolling Volatility ===
df_momentum['MomentumReturn'].rolling(window=12).std().plot(title='Rolling 12-Month Volatility')
plt.ylabel("Volatility (Std Dev)")
plt.grid(True)
plt.show()
```

## Strategy C: Momentum + Inverse Volatility

```python
combo_returns = []

for date in common_index[lookback_months:]:
    # 1. Select top industries by momentum
    ranked = momentum_scores.loc[date].sort_values(ascending=False)
    selected = ranked.head(top_n).index

    # 2. Inverse volatility weighting
    vol = volatility.loc[date, selected].replace(0, np.nan).dropna()
    inv_vol = 1 / vol
    weights = inv_vol / inv_vol.sum()

    # 3. Portfolio return
    ret = (weights * ind49.loc[date, weights.index]).sum()
    combo_returns.append((date, ret))

# === Convert to DataFrame ===
df_combo = pd.DataFrame(combo_returns, columns=['Date', 'StrategyReturn']).set_index('Date')

# === Add RF and Compute Sharpe ===
df_combo['RF'] = ff5.loc[df_combo.index, 'RF']
df_combo['ExcessReturn'] = df_combo['StrategyReturn'] - df_combo['RF']

sharpe_ratio = (
    df_combo['ExcessReturn'].mean() /
    df_combo['ExcessReturn'].std()
) * np.sqrt(12)
```

```python
# === Print Results ===
print(f"\n Strategy C: Momentum + Inverse Volatility")
print(f"Sharpe Ratio: {sharpe_ratio:.4f}")
print(f"Avg Monthly Return: {df_combo['StrategyReturn'].mean():.4%}")
print(f"Std Dev: {df_combo['StrategyReturn'].std():.4%}")
```

```python
# === Plot Cumulative Return ===
(1 + df_combo['StrategyReturn']).cumprod().plot(title='Strategy C: Momentum + Inverse Volatility')
plt.ylabel("Growth of $1")
plt.grid(True)
plt.show()
```
